```python
%matplotlib inline
from pylab import *
```

```python
x = np.arange(-10,10,0.001)
I = np.zeros(len(x))
I[np.where(np.abs(x)<1e-3)] = 1

sigma = 0.1
I = np.exp(-x*x/2/sigma**2)
I/= I.sum()
```

```python
plt.plot(x, I)
```

```python
V = np.fft.ifftshift(np.fft.fft(np.fft.fftshift(I)))
```

```python
u = np.fft.fftshift(np.fft.fftfreq(len(x), d=0.001))
```

```python
plt.plot(u,np.abs(V), color="red", label="Absolute")
plt.plot(u,np.real(V), color="blue", ls="--", label="Real Part")
plt.plot(u,np.imag(V), color="green", ls="--", label="Imag Part")
plt.xlim(-10,10)
plt.xlabel("Frequency")
plt.ylabel("Fourier Magunitude, Real part, Imag Part")
```

```python
plt.plot(u,np.angle(V, deg=True), color="Orange", ls="--", label="Phase")
plt.axhline(0, ls="-", color="gray")
plt.xlim(-10,10)
plt.ylim(-180,180)
plt.ylabel("Phase (deg)")
```
